# Exercise 6: Stats - Wind Statistics

```python
#1 
import pandas as pd
import datetime
```

```python
#3
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data'
data = pd.read_csv(url, sep = "\s+", parse_dates = [[0,1,2]]) 
data.head()
```

```python
#4 datetime assumes years are in the 2000's as rawdata only shows last 2 digits
def fix_year(n):
    year = n.year - 100
    return datetime.date(year, n.month, n.day)

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix_year)

data.head()

    
```

```python
#5
data["Yr_Mo_Dy"] = pd.to_datetime(data["Yr_Mo_Dy"])

data = data.set_index('Yr_Mo_Dy')

data.head()
```

```python
#6 
data.isnull().sum()
```

```python
#7

data.notnull().sum()
```

```python
#8 sum means of all locations and divide by the number of locations

data.mean().sum() / len(data.columns)
```

```python
#9 
loc_stats = data.describe()
loc_stats
```

```python
#10
day_stats = pd.DataFrame()
day_stats['min'] = data.min(axis = 1) # min
day_stats['max'] = data.max(axis = 1) # max 
day_stats['mean'] = data.mean(axis = 1) # mean
day_stats['std'] = data.std(axis = 1) # standard deviations

day_stats.head()
```

```python
#11
data.loc[data.index.month == 1].mean()
```

```python
#12 yearly
data.groupby(data.index.to_period('A')).mean()
```

```python
#13 monthly
data.groupby(data.index.to_period('M')).mean()
```

```python
#14 weekly
data.groupby(data.index.to_period('W')).mean()
```

```python
#15
weeks = data.resample('W').agg(['min','max','mean','std']) #weekly min,max
# for 52 weeks
weeks.loc[weeks.index[1:53], 'RPT':'MAL'].head()
```
